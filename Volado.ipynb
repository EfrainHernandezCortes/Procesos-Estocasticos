Ejercicio 2. Escriba una función que simule un volado.

```python
def volado():
  if r.random()>0.5:
    return "Aguila"
  else:
    return "Sol"
```

```python
volado()
```

```python
lv=[]
for i in range (10):
  lv.append(volado())
print(lv)
```

```python
lva=[]
lvs=[]
N=10**3
a=0
s=0
for i in range (1,N):
  if volado() == "Aguila":
    a+=1
    s=s
  else:
    a=a
    s+=1
  lva.append(a/i)
  lvs.append(s/i)
print(lva)
print(lvs)
```

```python
import matplotlib.pyplot as plt
plt.title('Probabilidad de Águila o Sol')
plt.plot(lva)
plt.plot(lvs)
plt.xlabel('Número de volados')
plt.ylabel('Probabilidad')
plt.legend(['Águila','Sol'])
plt.show()
```
