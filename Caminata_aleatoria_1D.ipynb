Se considera una caminata aleatoria en 1 dimensión, con probabilidad $p$ de moverse a la derecha y probabilidad $q$ de moverse a la izquierda, tal que $p + q = 1$.

Se importan las librerías necesarias.

```python
import random as r
import numpy as np
import matplotlib.pyplot as plt
```

Se considera una probabilidad aleatoria para $p$. Este valor también se puede elegir.


```python
pp = r.random()
pp
```

A través de la función _decision_ se asigna el valor correspondiente para determinar si avanza o retrocede una posición.

```python
def decision(pp):
  if r.random()>pp:
    return 1
  else:
    return -1
```

Se muestra la gráfica de la trayectoria de la partícula en los $n$ pasos.

```python
def trayectoria(pp):
  n = 10**2
  pasos = []
  dec=0
  caminata = [0]
  for i in range (n):
    pasos.append(decision(pp))
    dec+=pasos[i]
    caminata.append(dec)
  plt.title("Caminata aleatoria")
  plt.plot(caminata)
  plt.show
trayectoria(pp)
```

2. Para responder esta pregunta se debe verificar si hay al menos un $0$ en la lista _caminata_. Para ello, se corrige de la función _trayectoria_ de tal modo que la lista no comience en $0$ para no confundir con el punto incial.

```python
def trayectoria0(pp):
  n = 10**2
  pasos = []
  dec=0
  caminata = []
  for i in range (n):
    pasos.append(decision(pp))
    dec+=pasos[i]
    caminata.append(dec)
  return caminata

caminata0 = trayectoria0(pp)
if 0 in caminata0:
  print("Regresa al origen")
else:
  print("No regresa al origen")
```

3. Se calcula la posición promedio de la trayectoria.

```python
def trayectoriapromedio(pp):
  n = 10**2
  pasos = []
  dec=0
  caminata = []
  for i in range (n):
    pasos.append(decision(pp))
    dec+=pasos[i]
    caminata.append(dec)
  return caminata

caminata3 = trayectoriapromedio(pp)
promedio = np.average(caminata0)
print("La posición promedio es:", promedio)
```

4. Se revisa la posición dentro de la lista en la que se encuentra el $0$.

```python
def trayectoriatiempo(pp):
  n = 10**2
  pasos = []
  dec=0
  caminata = []
  for i in range (n):
    pasos.append(decision(pp))
    dec+=pasos[i]
    caminata.append(dec)
  return caminata

caminata4 = trayectoriatiempo(pp)
m = [i for i,x in enumerate(caminata4) if x==0]
print("El tiempo que le toma llegar al origen es:", m)
```

5. Se encuentra el lugar más lejano al origen al que llegó la partícula.

```python
def trayectorialejano(pp):
  n = 10**2
  pasos = []
  dec=0
  caminata = []
  for i in range (n):
    pasos.append(decision(pp))
    dec+=pasos[i]
    caminata.append(dec)
  return caminata

caminata5 = trayectorialejano(pp)
lejano = max(abs(x) for x in caminata5)
print("El punto más lejano al origen es:", lejano)
```
